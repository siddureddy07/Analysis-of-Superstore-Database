```python
# Preprocessing

# Handling missing data (if any)
print(df.isnull().sum())  # Check the number of missing values in each column

# If there are any missing values, you can decide how to handle them.
# For example, you might drop rows or columns with missing data or impute the missing values with mean/median.

```

```python
df.dropna(inplace=True)
```

```python
# Convert date columns to datetime data type
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract useful information from date columns (if needed)
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Ship Year'] = df['Ship Date'].dt.year
df['Ship Month'] = df['Ship Date'].dt.month

```

```python
plt.figure(figsize=(14, 6))

# Bar plot for Sales by Category
plt.subplot(1, 2, 1)
sns.barplot(x='Category', y='Sales', data=df, palette='pastel')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')

# Bar plot for Profit by Category
plt.subplot(1, 2, 2)
sns.barplot(x='Category', y='Profit', data=df, palette='pastel')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')


plt.tight_layout()
plt.show()
```

```python
# Visualization 2: Correlation Heatmap
correlation_matrix = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
plt.figure(figsize=(8, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

```

```python
plt.figure(figsize=(10,4))

# Bar plot for Sales by Region
plt.subplot(1, 2, 1)
sns.barplot(x='Region', y='Sales', data=df, palette='muted')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

# Bar plot for Profit by Region
plt.subplot(1, 2, 2)
sns.barplot(x='Region', y='Profit', data=df, palette='muted')
plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')

plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Profit by Ship Mode
plt.figure(figsize=(8, 6))
sns.barplot(x='Ship Mode', y='Profit', data=df, palette='muted')
plt.title('Profit by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.show()

```

```python
# Visualization 5: Distribution of Sales and Profit
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Sales'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Sales')

plt.subplot(1, 2, 2)
sns.histplot(df['Profit'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Profit')
plt.show()


```

```python
# Visualization 6: Sales and Profit trends over time
plt.figure(figsize=(10, 5))
sns.lineplot(x='Order Date', y='Sales', data=df, ci=None, color='skyblue')
sns.lineplot(x='Order Date', y='Profit', data=df, ci=None, color='salmon')
plt.title('Sales and Profit Trends over Time')
plt.xlabel('Order Date')
plt.ylabel('Amount')
plt.legend(['Sales', 'Profit'])
plt.show()
```

```python
# Statistical Analysis
# Additional Statistical Calculations

# Mean and Standard Deviation of Sales and Profit
sales_mean = df['Sales'].mean()
sales_std = df['Sales'].std()

profit_mean = df['Profit'].mean()
profit_std = df['Profit'].std()

print("Mean Sales: ", sales_mean)
print("Standard Deviation Sales: ", sales_std)

print("Mean Profit: ", profit_mean)
print("Standard Deviation Profit: ", profit_std)

# Hypothesis Testing (Example: T-test)
# Performing a t-test to compare the mean sales of two categories (Furniture and Office Supplies)
furniture_sales = df[df['Category'] == 'Furniture']['Sales']
office_supplies_sales = df[df['Category'] == 'Office Supplies']['Sales']

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(furniture_sales, office_supplies_sales)
print("T-Statistic: ", t_stat)
print("P-Value: ", p_value)

if p_value < 0.05:
    print("The difference in mean sales between Furniture and Office Supplies categories is statistically significant.")
else:
    print("There is no statistically significant difference in mean sales between Furniture and Office Supplies categories.")
```
